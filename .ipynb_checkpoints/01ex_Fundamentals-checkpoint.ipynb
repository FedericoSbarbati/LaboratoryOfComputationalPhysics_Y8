2\. Convert the following function into a pure function with no global variables or side effects

````python
x = 5
def f(alist):
    for i in range(x):
        alist.append(i)
    return alist

alist = [1,2,3]
ans = f(alist)
print (ans)
print (alist) # alist has been changed!
````

````python
import copy

x = 5
def f(alist):
    copy_list = copy.copy(alist)
    for i in range(x):
        copy_list.append(i)
    return copy_list

alist = [1,2,3]
ans = f(alist)
print (ans)
print (alist) # alist has been changed!
````

3\. Write a `decorator` hello that makes every wrapped function print “Hello!”, i.e. something like:

```python
@hello
def square(x):
    return x*x
```



````python
def hello(func):
    def wrapper(*args):
        print('Hello')
        return func(*args)
    return wrapper

@hello
def square(x):
    return x * x

print(square(5))
````

4\. Write the factorial function so that it a) does and b) does not use recursion.

````python
# Recursion

def factorial_rec(n):
    if(n-1 != 0):
        return n*(factorial_rec(n-1))
    else:
        return(n)

fac = factorial_rec(5)
print(fac)

# Not Recursion

print( 'No recursion')

def factorial(n):
    fac,b = n,1
    print(fac,b )
    for i in range (n-1):
        fac , b = fac*b, b+1
    return fac
    
fac = factorial(5)
print(fac)
````

5\. Use HOFs (zip in particular) to compute the weight of a circle, a disk and a sphere, assuming different radii and different densities:

```python
densities = {"Al":[0.5,1,2],"Fe":[3,4,5],"Pb": [15,20,30]}
radii = [1,2,3]
```

where the entries of the dictionary's values are the linear, superficial and volumetric densities of the materials respectively.

In particular define a list of three lambda functions using a comprehension that computes the circumference, the area and the volume for a given radius.


````python
# Calcolare pesi di un Cerchio, Cilindro omogeneo e Sfera. 

import math

densities = {"Al":[0.5,1,2],"Fe":[3,4,5],"Pb": [15,20,30]}
radii = [1,2,3]

# Function containing 3 lambda function to evaluate Circumference, Area and Volume
funcs = [
    (lambda r,density: 2*math.pi* r * density),
    (lambda r,density: math.pi* r**2 * density),
    (lambda r,density: (4/3)*math.pi* r**3 * density)
]

for i in radii:
    for key,value in densities.items():
        mass = [f(i,d) for f,d in zip(funcs,value)] 
        print(key, value , 'Ratio:' ,i , ' Mass:', mass)
    
````

````python
# Parent class
class Dog:

    # Class attribute
    species = 'mammal'

    # Initializer / Instance attributes
    def __init__(self, name, age, is_hungry):
        self.name = name
        self.age = age
        self.is_hungry = True

    # instance method
    def description(self):
        return "{} is {} years old".format(self.name, self.age)

    # instance method
    def speak(self, sound):
        return "{} says {}".format(self.name, sound)

    def eat(self):
        self.is_hungry = False

# Child class (inherits from Dog class)
class RussellTerrier(Dog):
    def run(self, speed):
        return "{} runs {}".format(self.name, speed)

# Child class (inherits from Dog class)
class Bulldog(Dog):
    def run(self, speed):
        return "{} runs {}".format(self.name, speed)


names = ['Tom','Fletcher','Larry']
ages = [6,7,9]
hungry = [True,True,True]

dog_list = [Dog(name,age,is_hungry) for name,age,is_hungry in zip(names,ages,hungry)]
for dog in dog_list:
    dog.eat()

print ('I have {} dogs'.format(len(dog_list)))
for dog in dog_list:
    print(dog.description())
print('And they are all {} of course'.format(dog.species))
check_all = True
for dog in dog_list:
    if dog.is_hungry == True:
        check_all = False
        print(dog.description())

if check_all:
    print('My dogs are not hungry.')
else:
    print('My dogs are hungry.')



````
