1\. Write a function that converts number representation, bin<->dec<->hex. (Clearly using the corresponding python built-in functions is not fair..)

```python
number = 30

def convert_to_base(number,base):
    if base not in [2,16]:
        print("Base sbagliata")
        return
    available_digits = [str(x) for x in range(10)] + [chr(i) for i in range(97, 123)]    
    digit = []
    while number >= 1: 
        carry = int(number % base)
        digit.append(available_digits[carry])
        
        number //= base # Divisione intera
    digit.reverse()
    return ''.join(digit)

print(convert_to_base(number,5))
binary = convert_to_base(number,2)
print(binary)
hexadecimal = convert_to_base(number,16)
print(hexadecimal)

    
```

2\. Write a function that converts a 32 bit word into a single precision floating point (i.e. interprets the various bits as sign, mantissa and exponent)

```python
def single_precision_FP(bit):

    if len(bit) != 32:
        return 'Codifica sbagliata'

    # Defining bit structure 
    sign_bit = 1
    mantissa_bit = 23
    exponent_bit = 8

    # Extracting binary rapresentation of sign,exponent,mantissa
    sign = bit[sign_bit-1]
    exponent = bit[sign_bit:exponent_bit+1]
    mantissa = bit[exponent_bit+1:]

    # Definining dictionary for structure
    IEE_standard = {"sign":sign,"exponent":exponent,"mantissa":mantissa}
    
    # Number interpretation
    exponent_is_zero = False
    exponent_is_max = False
    fraction_is_zero = False

    sign = 0
    exponent = 0
    mantissa = 0
    
    for key,value in IEE_standard.items():        
        if key == 'sign':
            sign = (-1)**(int(value))
            
        if key == 'exponent':
            # Decimal conversion
            decimal = 0
            for i,j in enumerate(value[::-1]):
                exponent += (int(j)*2**i)
            
        if key == 'mantissa':

            # Decimal conversion for fraction
            decimal = 0
            for i,j in enumerate(value):
                decimal+= (int(j)*2**(-i-1))
            mantissa = decimal

    if exponent == 0:
        exponent_is_zero = True
    if exponent == 255:
        exponent_is_max = True
    if mantissa == 0.0:
        fraction_is_zero = True
        
    # Normal definition
    if not exponent_is_zero and not exponent_is_max:
        return float(sign*(1.0+mantissa)*(2**(exponent-127)))

    if exponent_is_zero:

        # Positive and negative zero
        if fraction_is_zero:
            if sign == +1:
                return '+0'
            if sign == -1:
                return '-0'
        # Denormalized mantissa
        else:
            exponent = -126
            return sign*mantissa*(2**exponent)
    
    if exponent_is_max:

        if fraction_is_zero:
            if sign == +1:
                return '+inf'
            if sign == -1:
                return '-inf'
        else:
            return 'Nan'
            
        
            
        
```

```python
number = '10' # -5
print('Input number:',number)
decimal = single_precision_FP(number)
print(decimal)

number = '11000000101000000000000000000000' # -5
print('Input number:',number)
decimal = single_precision_FP(number)
print(decimal)

number = '01000000101000000000000000000000' # -5
print('Input number:',number)
decimal = single_precision_FP(number)
print(decimal)

number = '00000000000000000000000000000000' # +0
print('Input number:',number)
decimal = single_precision_FP(number)
print(decimal)

number = '10000000000000000000000000000000' # -0
print('Input number:',number)
decimal = single_precision_FP(number)
print(decimal)

number = '01111111100000000000000000000000' # +inf
print('Input number:',number)
decimal = single_precision_FP(number)
print(decimal)

number = '11111111100000000000000000000000' # -inf
print('Input number:',number)
decimal = single_precision_FP(number)
print(decimal)

number = '01111111110000000000000000000000' # Nan
print('Input number:',number)
decimal = single_precision_FP(number)
print(decimal)

number = '00111111100000000000000000000000'
print('Input number:',number)
decimal = single_precision_FP(number)
print(decimal)

number = '00000000010000000000000000000000'
print('Input number:',number)
decimal = single_precision_FP(number)
print(decimal)

```

3\. Write a program to determine the underflow and overflow limits (within a factor of 2) for python on your computer. 

**Tips**: define two variables inizialized to 1 and halve/double them enough time to exceed the under/over-flow limits  

```python
import math

massimo = float(1.0)
while not math.isinf(massimo*2):
    massimo*=2
print(massimo)
```

```python
minimo = float(1.0)
while (minimo/2) > 0:
    minimo/=2

print(minimo)
```

4\. Write a program to determine the machine precision

**Tips**: define a new variable by adding a smaller and smaller value (proceeding similarly to prob. 2) to an original variable and check the point where the two are the same 

```python
starter = 1.0
increment = 1.0

while (starter + increment)!= starter:
    increment/=2

print(increment)
    
```

5\. Write a function that takes in input three parameters $a$, $b$ and $c$ and prints out the two solutions to the quadratic equation $ax^2+bx+c=0$ using the standard formula:
$$
x=\frac{-b\pm\sqrt{b^2-4ac}}{2a}
$$

(a) use the program to compute the solution for $a=0.001$, $b=1000$ and $c=0.001$

(b) re-express the standard solution formula by multiplying top and bottom by $-b\mp\sqrt{b^2-4ac}$ and again find the solution for $a=0.001$, $b=1000$ and $c=0.001$. How does it compare with what previously obtained? Why?

(c) write a function that compute the roots of a quadratic equation accurately in all cases

```python
import math

a,b,c = 0.001,1000.0,0.001

delta = math.sqrt(math.pow(b,2)- 4*a*c)

x1 = (-b + delta) / (2*a)
x2 = (-b - delta) / (2*a)

print( 'x1:' , x1, 'x2:' , x2)

x1_1 = (2*c)/(-b-delta)
x2_2 = (2*c)/(-b+delta)

print( 'x1:' , x1_1, 'x2:' , x2_2)
```

```python
def square_algorithm(a,b,c):

    # IDEA: subtracting/adding number with different magnitude has low machine precision ( 10^-16)
    # while molticplication and division have higher precision

    delta = b**2 -4*a*c
    if delta < 0:
        print ('No real root')
        return
    if delta == 0:
        return -b/(2*a)

    if b>0:
        x1 = (-b - delta**0.5)/(2*a)
        x2 = c/(a*x1)

        return x1,x2

    if b<0: 
        x2 = (-b + delta**0.5)/(2*a)
        x1 = c/(a*x2)

        return x1,x2
```

```python
x1,x2 = square_algorithm(a,b,c)
print(x1,x2)
```

6\. Write a program that implements the function $f(x)=x(x−1)$

(a) Calculate the derivative of the function at the point $x = 1$ using the derivative definition:

$$
\frac{{\rm d}f}{{\rm d}x} = \lim_{\delta\to0} \frac{f(x+\delta)-f(x)}{\delta}
$$

with $\delta = 10^{−2}$. Calculate the true value of the same derivative analytically and compare with the answer your program gives. The two will not agree perfectly. Why not?

(b) Repeat the calculation for $\delta = 10^{−4}, 10^{−6}, 10^{−8}, 10^{−10}, 10^{−12}$ and $10^{−14}$. How does the accuracy scales with $\delta$?

```python
delta = 0.01
x = 1.0
f = lambda x: x*(x-1.0)
f1 = lambda x: 2.0*x -1.0


derivative = (f(x + delta)-f(x))/delta # 1 + delta , convergence to the limit for delta to zero
exact_derivative = f1(x)
print('Formula:',derivative)
print('Formula:', f(x+delta) , '  -  ' , f(x), '   /   ', delta) 
print('Exact result:',exact_derivative)
```

```python
delta_list = [10**-2, 10**-4,10**-6,10**-8,10**-10,10**-12,10**-14]
x = 1.0

f = lambda x: x*(x-1.0)

for delta in delta_list:
    derivative = (f(x + delta)-f(x))/delta # 1 + delta , convergence to the limit for delta to zero
    print(delta,' :',derivative)
```
