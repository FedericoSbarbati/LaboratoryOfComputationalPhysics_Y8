1\. **Kernel Density Estimate**

Produce a KDE for a given distribution (by hand, not using seaborn!):

* Fill a numpy array, x,  of len(N) (with N=O(100)) with a variable normally distributed, with a given mean a standard deviation
* Fill an histogram in pyplot taking properly care about the aesthetic
   * use a meaningful number of bins
   * set a proper y axis label
   * set proper value of y axis major ticks labels (e.g. you want to display only integer labels)
   * display the histograms as data points with errors (the error being the poisson uncertainty)
* for every element of x, create a gaussian with the mean corresponding the element value and std as a parameter that can be tuned. The std default value should be:
$$ 1.06 * x.std() * x.size ^{-\frac{1}{5.}} $$
you can use the scipy function `stats.norm()` for that.
* In a separate plot (to be placed beside the original histogram), plot all the gaussian functions so obtained
* Sum (with np.sum()) all the gaussian functions and normalize the result such that the integral matches the integral of the original histogram. For that you could use the `scipy.integrate.trapz()` method


````python
import numpy as np
import numpy.random as npr

# Data generation 

mean = 3
st_d = 0.5
size = 300
data = np.random.normal(loc=mean, scale=st_d, size=size)
print(data.shape)
````

````python
import matplotlib.pyplot as plt

N_bins = int(np.floor(np.sqrt(size)))

# Plotting data histogram

counts, bin_edges = np.histogram(data, bins = N_bins)

bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
errors = np.sqrt(counts)

print('N_bins: ', N_bins)
plt.hist(data, bins = N_bins)
plt.xlabel('Data measure')
plt.ylabel('Frequence')
plt.title('Hist')
plt.errorbar(bin_centers, counts, yerr = errors, fmt='o')

````

````python
from scipy.stats import norm

sigma = 1.06 * np.std(data) * (len(data)**(-1/5.))
gaussian_list = [norm(loc=xi, scale=sigma) for xi in data]
````

````python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 7))

# Data hist

ax1.hist(data, bins = N_bins)
ax1.set_xlabel('Data measure')
ax1.set_ylabel('Frequence')
ax1.set_title('Hist')
ax1.errorbar(bin_centers, counts, yerr = errors, fmt='o')

# Gaussian plot
T = np.linspace(data.min()-3*sigma, data.max()+3*sigma)

for gauss in gaussian_list:
    pdf_values = gauss.pdf(T)
    ax2.plot(T, pdf_values, color='gray', alpha=0.1, linewidth=1)
````

````python
from numpy import trapezoid

KDE_sums = np.sum([g.pdf(T) for g in gaussian_list],axis = 0)
target_integral = len(data)


# Calcola l'area attuale sotto la curva KDE (non ancora normalizzata)
current_integral = trapezoid(KDE_sums, T)

# Calcola il fattore di normalizzazione
norm_factor = current_integral / target_integral

# Normalizza la curva finale
KDE_normalized = KDE_sums / norm_factor

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 7))

# Data hist

ax1.hist(data, bins = N_bins)
ax1.set_xlabel('Data measure')
ax1.set_ylabel('Frequence')
ax1.set_title('Hist')
ax1.errorbar(bin_centers, counts, yerr = errors, fmt='o')

# Gaussian plot
T = np.linspace(data.min()-3*sigma, data.max()+3*sigma)


````
